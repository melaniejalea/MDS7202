1. Control de versiones es un sistema que registra los cambios a documentos o conjuntos de archivos, lo que permite verificar estados previos cuando sea necesario. De esta manera, si algún archivo, código u otros elementos fueron eliminados accidentalmente o ya no funcionan, se puede acceder a ellos haciendo un "backtracking" hasta una versión donde estaban presentes y/o funcionales.

`Git` es un sistema de control de versiones **distribuido** moderno que permite hacer un seguimiento del historial de archivos. Este sistema se diferencia de un control de versiones **centralizado** en los siguientes puntos:

- Los clientes no solo pueden descargar la última copia instantánea de los archivos, sino que se replica completamente el repositorio.
- Cada clon es realmente una copia completa de todos los datos. 
- Permite establecer varios flujos de trabajo que no son posibles en sistemas centralizados


2. 
- **Primera diferencia**: `Git` corresponde a un sistema de control de versiones para verificar cambios realizados en archivos y `Github` corresponde a una interfaz gráfica para utilizar este sistema.

- **Segunda diferencia** : `Git` es una herramienta de línea de comando  y `Github` es un plataforma de servicio online de proyectos que usan `Git`. 

- **Tercera diferencia** : `Git` crea un repositorio local para realizar un seguimiento de los cambios locales en el código fuente, mientras que `Github` es un servicio en línea que permite subir repositorios locales para almacenarlos en un repositorio remoto en la nube.

3. Un <code>hash</code> corresponde a un identificador único de 40 caracteres que sirven para identificar los <code>commits</code> realizados dentro de un repositorio. Estos se pueden verificar haciendo uso del comando <code>git log</code>. Un ejemplo de esto corresponde al siguiente:

 >   Commit de ejemplo a

 En donde los identificadores de "d4f2aeb754e2f0e5794f7d2c66a8c3a61b4a4f65" y "8bffd3bfcf2b52b3e119a74ef2ac69a9d33b10d1" corresponden a los <code>hash</code>.


4. Los elementos que componen la estructura de datos de <code>Git</code> corresponden a:
 - `Blob` : Tipo de objeto que se utiliza para almacenar el contenido de cada archivo en un repositorio.
 - `Tree` : Tipo de objeto que crea la jerarquía entres los archivos para un repositorio de `Git`. Las hojas de estos árboles pueden corresponder tanto a <code>blobs</code> como a otros árboles.
 - `Commit` : Tipo de objeto que es una captura de la jerarquía y de los contenidos de los archivos en un repositorio. 
 - `Reference` : Puntero a un <code>hash</code> de un <code>commit</code> en específico del repositorio. 
 - `Tags` : Similar a <code>reference</code>, se utilizan para recordar un <code>commit</code> de alta relevancia para un proyecto, como versiones funcionales de código y parecidos.

11. `git stash` (0.25 puntos)
- Es un comando que se utiliza para poder almacenar de manera temporal los cambios en documentos/archivos sin enviarlos al repositorio, es decir, que aún no se ha realizado un <code>commit</code>. Sirve para poder trabajar en otra rama y luego regresar al trabajo inicial donde se almacenaron los cambios iniciados.
12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)
- Como se explicó en la sección anterior, <code>git revert</code> se utiliza para deshacer un <code>commit</code> previo sin afectar al repositorio compartido, y realiza esto creando un nuevo <code>commit</code>. En cambio, el comando `git reset` deshace cambios en el repositorio modificando su historial, cambiando el historial para las personas que también trabajan sobre el repositorio. Es un comando útil de utilizar cuando aún no se ha realizado un <code>commit</code>. La diferencia radica en como afectan al historial.

8. Cree una rama nueva y proponga una solución al problema anterior, adjuntando el código corregido y señalando las líneas modificadas. (0.25 pts)

Para crear una nueva rama después de que el proceso de fusión haya fallado, es necesario utilizar el comando <code>git merge --abort</code> para cancelar el intento de fusión. Git nos solicita resolver los conflictos antes de cambiar de rama. Si no ejecutamos este comando, no podremos regresar a la rama principal (main) para crear una nueva rama. 

Después de abortar el merge, creamos una nueva rama con el comando <code>git branch fix-merge</code>y nos cambiamos a esa rama de trabajo con <code>git checkout fix-merge</code>. Manteniendo la modularización, corregimos los condicionales <code>if</code> en la función animales del módulo animales, utilizando condicionales <code>elif</code>, luego podríamos commitear los cambios para posteriormente hacer un <code>merge</code> de <code>fix-merge</code> a <code>main</code>.

````python
#### código modificado

def animales(input):
    
    '''
    Como hace el animalito?
    '''
    
    if input.lower() == 'gato':
        print('miau')
    ####### línea original : if input.lower() == 'perro':
    elif input.lower() == 'perro': ### modificación!
        print('guau')
    ####### línea original : if input.lower() == 'fox':
    elif input.lower() == 'fox': ### modificación!
        print('https://www.youtube.com/watch?v=jofNR_WkoCE')
    else:
        raise ValueError('animal no reconocido :(')
````
